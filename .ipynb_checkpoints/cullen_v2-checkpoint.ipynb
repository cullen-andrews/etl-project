```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats as stats
from sqlalchemy import create_engine
```

```python
# Create dataframes for both the movie and the streaming data sources
# NOTE: The movie data source will cause a Warning because one record in the Year column contains text and the year. This
# will be resolved through the data cleaning process.

imdb_df = pd.read_csv("Resources/IMDB_movies.csv")
streaming_df = pd.read_csv("Resources/MoviesOnStreamingPlatforms_updated.csv")
```

```python
# Create dataframe with just the columns we want
imdb_df = pd.DataFrame(imdb_df[["title","genre","description","director","actors","year","duration","budget",\
                                "usa_gross_income","worlwide_gross_income","votes","metascore"]])
imdb_df
```

```python
# Remove the row with a non-year value in the year column
imdb_movies = imdb_df[imdb_df.year != "TV Movie 2019"]
imdb_int_df = pd.DataFrame(imdb_movies)
imdb_int_df
```

```python
# Fix the spelling for the worlwide_gross_income column
imdb_int_df = imdb_int_df.rename(columns= {"worlwide_gross_income": "worldwide_gross_income"})
imdb_int_df
```

```python
# Change the year column to int and check data types
imdb_int_df["year"] = imdb_int_df["year"].astype(int)

imdb_int_df.dtypes
```

```python
# Remove any rows in the dataframe in which there is no director or no actor
imdb_int_df = imdb_df.dropna(subset=["director", "actors"])
imdb_movie_df = pd.DataFrame(imdb_int_df)
imdb_movie_df
```

```python
# Combine year and title into one column to use as the unique identifier (primary key for the database table)
year_title=imdb_movie_df["year"].astype(str) + "_" + imdb_movie_df["title"]
imdb_movie_df["year_title"] =  year_title
imdb_movie_df
```

```python
#imdb_split_list = imdb_df.Actors.str.split(',').tolist()

# Create table of each movie and each actor in each of those movies (junction table for the database)
# Split each actor into own row when multiple actors in one movie
imdb_stack_df = pd.DataFrame(imdb_movie_df.actors.str.split(',').tolist(), index=imdb_movie_df.year_title).stack().reset_index()
imdb_stack_df
```

```python
# Remove the inner index, level_1
imdb_stack_df = imdb_stack_df.drop(columns="level_1")
imdb_stack_df
```

```python
# Rename column headers
imdb_stack_df.columns = ["year_title", "Actor"]
imdb_stack_df
```
